```python
# 33. Display Sex wise survival status of passengers who paid more fare. 
#       A. Above Average.
#       B. Above III Quartile. [>= 75 Percentile]
#       C. Top 10%.


# A. Above Average
fare_avg = df['Fare'].mean()
above_avg = df[df['Fare'] > fare_avg].groupby(['Sex', 'Survived'])['PassengerId'].count()
print("Above Average Fare:\n", above_avg)

# B. Above III Quartile (>= 75th Percentile)
fare_75 = df['Fare'].quantile(0.75)
above_75 = df[df['Fare'] >= fare_75].groupby(['Sex', 'Survived'])['PassengerId'].count()
print("Above 75th Percentile Fare:\n", above_75)

# C. Top 10% 
fare_90 = df['Fare'].quantile(0.90)
top_10 = df[df['Fare'] >= fare_90].groupby(['Sex', 'Survived'])['PassengerId'].count()
print("Top 10% Fare:\n", top_10)
```

```python
# Create a new column family_size by adding sibsp (siblings/spouses) and parch (parents/children) together.

df['family_size'] = df['SibSp'] + df['Parch']
df['family_size']
```

```python
# Find the survival rate of passengers who traveled alone (family_size = 0).
alone_passengers = df[df['family_size'] == 0]

# Calculate the survival rate
survival_rate_alone = alone_passengers['Survived'].mean()
survival_rate_alone
```

```python
# Find the probability of survival for passengers aged between 20 and 30.
young_adults = df[(df['Age'] >= 20) & (df['Age'] <= 30)]

# Calculate the survival rate
survival_rate_of_Young_Adults = young_adults['Survived'].mean()
survival_rate_of_Young_Adults
```
