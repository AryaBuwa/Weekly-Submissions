### Machine Learning 
#### Regression Methods 
        1. Logistic Regression 
        - Logistic Regression is a supervised machine learning algorithm used for classification tasks where the goal is to predict the probabilty that an instance belongs to a given class or not.
        - It is a binary classification algorithm, meaning it can only classify instances into two distinct classes.
        - The algorithm assumes that the relationship between the input features and the target variable follows a logistic function.
        - The main assumptions made by logistic regression include:
        - Independence of observations: The observations are independent of each other.
        - Linearity of the independent variables: The independent variables must be linearly related to the log odds of the target variable.
        - No multicollinearity: The independent variables should not be highly correlated with each other.
        - Lack of perfect separation: The target variable should not perfectly separate the instances into the two classes.
        - Homoscedasticity: The variance of the residuals is constant across the entire dataset.
        - Logistic Regression can be extended to handle more than two classes using techniques like one-vs-rest or multinomial logistic regression.
        - Logistic Regression is a popular choice for classification tasks due to its simplicity, interpretability, and ease of implementation. It is widely used in various fields, including medicine, finance, and marketing.

```python
# importing necessary libraries

import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib
```

```python
# Loading the Dataset
df = pd.read_csv(r'C:\Users\User\Downloads\Bigmart.csv')
df
```

```python
# Performing Data Preprocessing
# Removing the Null Values
df.dropna(inplace=True) 
```

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
# Converting the categorical data into numerical values using Label Encoding
from sklearn.preprocessing import LabelEncoder

label_encoder = {}
    
for col in categorical_cols:
    try :
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        label_encoder[col] = le
    except Exception as e :
        print(f"Error occurred while encoding column {col}: {str(e)}")
```

```python
# Define features and target variable 
if "Item_MRP" in df.columns:
    x = df.drop('Item_MRP', axis=1)
    y = df['Item_MRP']
else :
    raise KeyError("Column 'Item_MRP' not found in the dataset") 
```

```python
# Checking for Outliers

plt.figure(figsize=(10, 5))
plt.specgram(df['Item_MRP'], NFFT=256, Fs=100, cmap='plasma', scale='dB')
plt.xlabel("Time", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Item MRP Spectrum Analysis", fontsize=14, fontweight='bold')
plt.colorbar(label='Intensity (dB)')
plt.show()
```

```python
# Removing Outliers

Q1 = df['Item_MRP'].quantile(0.25)
Q3 = df['Item_MRP'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Item_MRP'] >= lower_bound) & (df['Item_MRP'] <= upper_bound)]

```

```python
# Checking for Multicollinearity

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
```

```python
# Feature Selection and Encoding
categorical_cols = ['Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Defining the features and target variable

x = df.drop('Item_MRP', axis=1)
y = df['Item_MRP']
```

```python
# Train Test and Split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
# Model Training
model = LinearRegression()
LR = LinearRegression
```

```python
model.fit(x_train, y_train)
```

```python
# Prediction of the model

y_pred = model.predict(x_test)
```

```python
# Predicted Values Distribution Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(y_pred, fill=True, color='green', alpha=0.6)
plt.xlabel("Predicted Item MRP", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.title("Distribution of Predicted MRP", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Model Evaluation
# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# R-squared

r2 = r2_score(y_test, y_pred)
```

```python
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```
