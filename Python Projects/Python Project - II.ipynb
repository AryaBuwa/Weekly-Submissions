```python
# define the TO-DO List class

class todolist:
    def __init__(self):
        self.tasks = []   # list of tasks

    def add_task(self, task):
        # Append the task as a dictionary.
        self.tasks.append({"Task":task, "Completed": False})
    
    def remove_task(self, task_index):

        # check if the task_index is valid
        if 0 <= task_index < len(self.tasks):
            # removing the task and printing the removal message.
            removed_task = self.tasks.pop(task_index)
            print(f"Task '{removed_task['task']}' removed.")
        else : 
            # If the index is invalid, print an error message.
            print("Invalid task index!")

    def mark_completed(self, task_index):
        # checking if the task index is valid or not
        if 0 <= task_index < len(self.tasks):
            # Mark the task as completed
            self.tasks[task_index]["Completed"] = True
            print(f"Task '{self.tasks[task_index]['task']}' marked as completed.")
        else : 
            print("Invalid task index!")

    def show_tasks(self):
        # Check if there are any tasks
        if not self.tasks:
            print("No tasks to display ")
            return

        # print each task with its status (completed or pending)
        for idx, task in enumerate(self.tasks):
            status = "Completed" if task["completed"] else "Pending"
            print(f"{idx}. {task['task']} [{status}]")


# Now Writing a Program to interact with TO-DO List

def main():
    # creating an instance of todolist class
    todo_list = todolist() 

    while True:
        # Display the menu options for the user
        print("\nTo-Do List Menu :")
        print("1. Add Task")
        print("2. Remove Task")
        print("3. Mark Task as Completed")
        print("4. View Tasks")
        print("5. Exit")

        # Get the user input
        choice = input("Enter your choice : ")


        if choice == '1':
            # Asking user for a task 
            task = input("Enter a Task: ")
            todo_list.add_task(task) # adding the task to the list
            print('Task',task,'added.')

        elif choice == '2':
            # Show the current list of tasks
            todo_list.show_tasks()
            # ask the user for the task index to remove
            task_index = int(input("Enter th task index to remove: "))
            todo_list.remove_task(task_index) # Removing the task at given index

        elif choice == "4":
            # Show the current list of tasks
            todo_list.show_tasks()

        elif choice == '5':
            # Exit the Program
            print("Goodbye!")
            break
        
        else :
            # If user enters an invalid option, print an error message.
            print("Invalid choice! Please select a valid option.")

# Running the final program to check if it works or not
if __name__ == "__main__":
    main()
```
